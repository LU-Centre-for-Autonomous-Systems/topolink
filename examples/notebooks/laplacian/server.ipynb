# Test for consensus

This notebook documents the server-side implementation for the consensus test.
Here, we define a ring topology with $5$ nodes.

```python
def server() -> None:
    from logging import basicConfig, INFO
    import matplotlib.pyplot as plt

    basicConfig(level=INFO)

    from topolink import Graph

    nodes = ["1", "2", "3", "4", "5"]
    edges = [("1", "2"), ("2", "3"), ("3", "4"), ("4", "5"), ("5", "1")]

    graph = Graph(nodes, edges, address="localhost:5555")

    _, ax = plt.subplots()

    graph.draw(ax)

    plt.show()

    graph.deploy()


server()
```
